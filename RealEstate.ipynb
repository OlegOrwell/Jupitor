```python
df_train['Healthcare_1'] = df_train['Healthcare_1'].fillna(df_train["Healthcare_1"].mean())
df_train['LifeSquare'] = df_train['LifeSquare'].fillna(df_train["LifeSquare"].mean())
df_test['Healthcare_1'] = df_train['Healthcare_1'].fillna(df_train["Healthcare_1"].mean())
df_test['LifeSquare'] = df_train['LifeSquare'].fillna(df_train["LifeSquare"].mean())
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(df_train.drop(['Price', "Id"], axis = 1), df_train['Price'], test_size = 0.15)
#X_train, X_valid, y_train, y_valid = train_test_split(df_train.drop(['Price'], axis = 1), df_train['Price'], test_size = 0.15)
```

```python
X_train_scale = standard_scaler.fit_transform(X_train)
X_valid_scale = standard_scaler.transform(X_valid)
```
